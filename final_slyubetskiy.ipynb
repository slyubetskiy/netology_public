# 7. Проверить гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом

```python
df['salary'].value_counts()
```

```python
df.groupby(['salary'])[['average_montly_hours']].mean()
```

среднее время на работе почти одинаковое для всех уровней оклада

```python
data_1 = df.loc[df['salary']=='high']['average_montly_hours'].values
data_2 = df.loc[df['salary']=='low']['average_montly_hours'].values
```

```python
import seaborn as sns
```

```python
sns.kdeplot(data_1, label="high salary", fill = True, color = '#000000')
sns.kdeplot(data_2, label="low salary", fill = True, color = '#B22222')
plt.legend()
plt.show()
```

```python
import scipy.stats as st
```

```python
stat, p = st.shapiro(data_1)

print(f'stat = {stat:.3f}, p = {p:.5f}')
if p > 0.05:
    print('Вероятно нормальное распределение')
else:
    print('Вероятно не нормальное распределение')
```

```python
stat, p = st.shapiro(data_2)

print(f'stat = {stat:.3f}, p = {p:.5f}')
if p > 0.05:
    print('Вероятно нормальное распределение')
else:
    print('Вероятно не нормальное распределение')
```

```python
stat, p = st.mannwhitneyu(data_1, data_2)

print(f"Статистика = {stat:.5f}, p = {p:.5f}")

if p > 0.05:
    print('Не отклоняем нулевую гипотезу, распределения, вероятно, одинаковые')
else:
    print('Отклоняем нулевую гипотезу, распределения, вероятно, различаются')
```

<b>Вывод.</b> Тест не показал наличие статистически значимой разницы между временем, проведенным на работе и окладом

# 9. Разделить данные на тестовую и обучающую выборки.Построить модель LDA, предсказывающую уволился ли сотрудник на основе имеющихся факторов (кроме department и salary). Оценить качество модели на тестовой выборке

```python
df.head()
```

```python
# импортируем LDA-функционал
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
X = df[['satisfaction_level'	,'last_evaluation'	,'number_project',	'average_montly_hours',	'time_spend_company'	,'Work_accident','promotion_last_5years']]
y = df['left']
```

```python
# разделяем выборку на тренировочную и тестовую
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
# обучаем данные
lda.fit(X_train, y_train)
```

```python
# делаем прогноз на тестовой выборке
lda.predict(X_test)
```

```python
# расчет точности
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, lda.predict(X_test))
```
